```python
##### CC video #####
##### CC video #####
##### CC video #####

path = "/Users/liyuhan/python_practice/NetworkCons/CCsamplevideo"
file_names = []
for file in os.listdir(path):
    file_path = os.path.join(path, file)
    if file_path.endswith('.csv'):
        file_names.append(file_path)
    
    for filename in file_names:
        if not filename.endswith('.csv'):
            continue
        print(file_names)
            
    mydf = pd.read_csv(filename)
    mydf = pd.DataFrame(mydf)
    mydf.insert(21, 'reply_user', '')
    
    match = mydf['回复内容'].str.findall(r'回复 @(.*?)  :')
    mydf['reply_user'] = match
    
    mydf['reply_user'] = mydf['reply_user'].fillna('')
    mydf['回复作者'] = mydf['回复作者'].fillna('')
    mydf['评论作者'] = mydf['评论作者'].fillna('')
    
    reply_to_user_list = []
    for index, row in mydf.iterrows():
        if len(mydf['reply_user'][index]) > 0:
            if mydf['回复作者'][index] != '':
                reply_network = mydf['回复作者'][index] + ',' + ''.join(mydf['reply_user'][index])
            reply_to_user_list.append(reply_network)
        else:
            if mydf['回复作者'][index] != '':
                reply_network = mydf['回复作者'][index] + ',' + mydf['评论作者'][index]
            else:
                reply_network = mydf['评论作者'][index] + ','
            reply_to_user_list.append(reply_network)

    mydf2 = pd.DataFrame(reply_to_user_list, columns = ['replyer/reply_to'])
    mydf2[['replyer','reply_to']] = mydf2['replyer/reply_to'].str.split(',', expand=True)
    mydf2 = mydf2.drop('replyer/reply_to',axis=1)
    
    ### construct an adjacency matrix by using relationship csv ###
    list = mydf2.values.tolist()
    
    data = []
    for edgelist in list:
        data.append(edgelist)
    
    relation_list = data
    
    member_dict = {}
    member_index = 0
    for name_tuple in relation_list:
        for name in name_tuple:
            if name in member_dict:
                continue
            member_dict[name] = member_index
            member_index += 1
            
    relation_matrix = [[0 for i in range(len(member_dict))]
                       for i in range(len(member_dict))]
    
    for (x, y) in relation_list:
        x_index = member_dict[x]
        y_index = member_dict[y]
        relation_matrix[x_index][y_index] = 1
    
    matrix_df = pd.DataFrame(relation_matrix)
    matrix_df.to_csv(filename + '.csv')
```
